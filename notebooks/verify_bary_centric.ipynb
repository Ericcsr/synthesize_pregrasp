```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([0.1,0.2])
B = np.array([1.1,0.2])
C = np.array([0.6,0.9])
```

```python
plt.scatter([A[0],B[0],C[0]],[A[1],B[1],C[1]])
```

```python
points_x = []
points_y = []
for x in np.linspace(0,1,50):
    for y in np.linspace(0,1,50):
        x_ = 1-np.sqrt(x)
        y_ = np.sqrt(x)*(1-np.sqrt(y))
        z_ = np.sqrt(y*x)
        points_x.append(A[0]*x_+B[0]*y_+C[0]*z_)
        points_y.append(A[1]*x_+B[1]*y_+C[1]*z_)
plt.scatter(points_x,points_y,marker=".")
```
